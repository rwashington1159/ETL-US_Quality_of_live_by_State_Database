```python
import pandas as pd
```

```python
# Path
sales_df = "Sale_Prices_State.csv"
```

```python
# Read Sales csv and store into Pandas DF
sales_df = pd.read_csv(sales_df, encoding="ISO-8859-1", index_col=[0])
sales_df
```

Transform

```python
sales_df = sales_df.dropna(how='any')
sales_df
```

```python
# Rename Region Name column to State
sales_df = sales_df.rename(columns={"RegionName": "State",})
sales_df  
```

```python
med_sales_df = sales_df[['State', 'Median Sales Price']]
med_sales_df
```
